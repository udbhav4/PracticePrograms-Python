```python
import turtle
import time
import random


# Developing screen

win=turtle.Screen()
win.title("SNAKE AND DOTS                                                            - by Udbhav")
win.bgcolor("black")
win.setup(width=600, height=600)
win.tracer(0)

# Making snake head

head=turtle.Turtle()
head.speed(0)
head.shape("circle")
head.color("blue")
head.penup()
head.goto(0,0)
head.direction="stop"

# Making the dots for the snake

food=turtle.Turtle()
food.speed(0)
food.shape("triangle")
food.color("yellow")
food.penup()
x=random.randint(-280,280)
y=random.randint(-280,250)
food.goto(x,y)

body=[]
score=0
hscore=0
counter=0

# Printing the details of the game on the screen

pen=turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("Score:  {}".format(score), align="center", font=("Courier",18,"normal"))

# Tasks that go on in the game

def go_up():
    global counter
    counter=0
    if head.direction!="down": 
        head.direction="up"
def go_down():
    global counter
    counter=0
    if head.direction!="up":
        head.direction="down"
def go_right():
    global counter
    counter=0
    if head.direction!="left":
        head.direction="right"
def go_left():
    global counter
    counter=0
    if head.direction!="right":
        head.direction="left"

def move():
    if head.direction=="up":
        head.sety(head.ycor()+0.25)
        shift_up()
        
    if head.direction=="down":
        head.sety(head.ycor()-0.25)
        shift_down()
        
    if head.direction=="right":
        head.setx(head.xcor()+0.25)
        shift_right()
        
    if head.direction=="left":
        head.setx(head.xcor()-0.25)
        shift_left()

# Moving the body parts of the snake.
        
def shift_up():
    global counter
    for i in range(len(body)-1,0,-1):
        x=body[i-1].xcor()
        y=body[i-1].ycor()-10
        body[i].goto(x,y)
        if counter==0:
            time.sleep(0.07)
    counter=1
    # We are moving the last body part to its immediate above/older body part. By doing this, all the rest of the body parts, except the first added body part, go to the place of the upper body part.
    
    # Moving the first added body part in the place of the head, 'if any body part is added.
    if len(body)>0:
        x=head.xcor()
        y=head.ycor()-10
        body[0].goto(x,y)

def shift_down():
    global counter
    for i in range(len(body)-1,0,-1):
        x=body[i-1].xcor()
        y=body[i-1].ycor()+10
        body[i].goto(x,y)
        if counter==0:
            time.sleep(0.07)
    counter=1
    # We are moving the last body part to its immediate above/older body part. By doing this, all the rest of the body parts, except the first added body part, go to the place of the upper body part.
    
    # Moving the first added body part in the place of the head, 'if any body part is added.
    if len(body)>0:
        x=head.xcor()
        y=head.ycor()+10
        body[0].goto(x,y)

def shift_right():
    global counter
    for i in range(len(body)-1,0,-1):
        x=body[i-1].xcor()-10
        y=body[i-1].ycor()
        body[i].goto(x,y)
        if counter==0:
            time.sleep(0.07)
    counter=1
    # We are moving the last body part to its immediate above/older body part. By doing this, all the rest of the body parts, except the first added body part, go to the place of the upper body part.
    
    # Moving the first added body part in the place of the head, 'if any body part is added.
    if len(body)>0:
        x=head.xcor()-10
        y=head.ycor()
        body[0].goto(x,y)
        
def shift_left():
    global counter
    for i in range(len(body)-1,0,-1):
        x=body[i-1].xcor()+10
        y=body[i-1].ycor()
        body[i].goto(x,y)
        if counter==0:
            time.sleep(0.07)
    counter=1
    # We are moving the last body part to its immediate above/older body part. By doing this, all the rest of the body parts, except the first added body part, go to the place of the upper body part.

    # Moving the first added body part in the place of the head, 'if any body part is added.
    if len(body)>0:
        x=head.xcor()+10
        y=head.ycor()
        body[0].goto(x,y)

        
# Keyboard bindings

win.listen()
win.onkeypress(go_up,"Up")
win.onkeypress(go_down,"Down")
win.onkeypress(go_right,"Right")
win.onkeypress(go_left,"Left")

# Making the game go continuously

while True:
    win.update()
    # Condition when the snake reaches the border
    if len(body)>0:
        x=body[len(body)-1].xcor()
        y=body[len(body)-1].ycor()
        if (x==290 and head.direction=="right") or (x==-290 and head.direction=="left"):
            head.setx(-x)
        if (y==290 and head.direction=="up") or (y==-290 and head.direction=="down"):
            head.sety(-y)
    else:
        x=head.xcor()
        y=head.ycor()
        if (x==290 and head.direction=="right") or (x==-290 and head.direction=="left"):
            head.setx(-x)
        if (y==290 and head.direction=="up") or (y==-290 and head.direction=="down"):
            head.sety(-y)

        
    # Condition when the food is found by the snake
    if head.distance(food)<20:
        # Moving food to random place
        x=random.randint(-280,280)
        y=random.randint(-280,250)
        food.goto(x,y)
        # Attaching another body part
        new=turtle.Turtle()
        new.speed(0)
        new.shape("circle")
        new.color("blue")
        new.penup()
        body.append(new)
        # Setting score and high score, and printing score
        score+=1
        if score>hscore:
            hscore=score
        pen.clear()
        pen.write("Score:  {}".format(score), align="center", font=("Courier",18,"normal"))
        
    # Checking for the collision of snake with its own body
    for part in body:
        if head.distance(part)<10:
            time.sleep(0.5)
            # Sending head to the centre to restart the game
            head.goto(0,0)
            head.direction="stop"
            # Hiding all the body parts somewhere out of the sight
            for part in body:
                part.goto(1000,1000)
            # Clearing the body parts list
            body.clear()
            # Printing the present score and the high score
            pen.clear()
            pen.write("Last Score:  {}    High Score:  {}".format(score,hscore), align="center", font=("Courier",18,"normal"))
            score=0
            time.sleep(3)
            pen.clear()
            pen.write("Score:  {}".format(score), align="center", font=("Courier",18,"normal"))
            
    move()
    #time.sleep(0.00000000000000000000000001)

win.mainloop()
```
